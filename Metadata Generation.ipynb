```python
#form metadata generation

master_xml_directory = input('Enter absolute directory of xml master: ')
master_xml = open(master_xml_directory,"r",encoding='utf-8')

metadata_creators = []
new_titles = []
countries = []
identifiers = []
dig_id = []
dates = []
continents = []
states = []
cities = []
coordinates = []
eras = []
authors = []
abstracts = []
languages = []
subjects = []
sources = []
conditions = []

project = "National Endowment for the Humanities Collaborative Research Grant"
oscar_project = "British Library EAP060"
publisher = "Slave Societies Digital Archive"
director = "Jane Landers(project director)"
cuba_director = "Paul Lovejoy(project director)"
brazil_director = "Mariza Soares(project director)"
oscar_director = "Oscar Grandío Moráguez (project director)"
types = ["Text","Ecclesiastical records"]
rights = "Vanderbilt University makes these historical records available for the limited purposes of individual private study, scholarship, research, teaching, and other educational or instructional uses. No rights of any kind are granted to users with respect to photographs or graphic images published on this site."

for line in master_xml:
    line_content = line[line.find(">") + 1: line.rfind("<")]   
    
    if "<Your_Name_>" in line:
        metadata_creators.append(line_content)
    elif "<New__generated_by_you__title_of_volume_>" in line:
        new_titles.append(line_content)
    elif "<Country_>" in line:
        countries.append(line_content)
    elif "<Volume_PID_>" in line:
        identifiers.append(line_content)
        poss_id = line_content[line_content.find(":") + 1:]
        if poss_id.isdigit():
            dig_id.append(int(poss_id))
        else:
            dig_id.append(line_content)        
    elif "<Date_s__>" in line:
        dates.append(line_content)
    elif "<Continent_of_origin_>" in line:
        continents.append(line_content)
    elif "<Region__State_Province__>" in line:
        states.append(line_content)
    elif "<City_>" in line:
        cities.append(line_content)
    elif "<Cartographic_coordinates_>" in line:
        coordinates.append(line_content)
    elif "<Time_period_>" in line:
        eras.append(line_content)
    elif "<Creator_>" in line:
        authors.append(line_content)
    elif "<Creator_/>" in line:
        authors.append("null")
    elif "<Abstract_>" in line:
        abstracts.append(line_content)
    elif "<Abstract_/>" in line:
        abstracts.append("null")
    elif "<Language_>" in line:
        languages.append(line_content)
    elif "<Contents_>" in line:
        subjects.append(line_content)
    elif "<Institution_>" in line:
        sources.append(line_content)
    elif "<Condition_>" in line:
        conditions.append(line_content)
    elif "<Condition_/>" in line:
        conditions.append("null")
        
master_xml.close()
current_volume = 0

while current_volume < len(new_titles):
    is_oscar = False
    if isinstance(dig_id[current_volume], int):
        if (countries[current_volume] == 'Cuba') and (153874 <= dig_id[current_volume] <= 178832):
            is_oscar = True
        elif (countries[current_volume] == 'Cuba') and (dig_id[current_volume] >= 203000) and (dig_id[current_volume] != 241315):
            is_oscar = True
    
    current_folder = "0" * (4 - len(str(current_volume + 1))) + str(current_volume + 1)
    print("Now working on " + new_titles[current_volume])
    current_dc = open("C:\\Users\\genkindn\\Form_metadata\\" + current_folder + "DC.xml","w",encoding = "utf-8")
    mods = open("C:\\Users\\genkindn\\Form_metadata\\" + current_folder + "MODS.xml","w",encoding = "utf-8")
        
    current_dc.write("<oai_dc:dc xmlns:dc=\"http://purl.org/dc/elements/1.1/\" xmlns:oai_dc=\"http://www.openarchives.org/OAI/2.0/oai_dc/\" xmlns:xsi=\"http://www.w3.org/2001/XMLSchema-instance\" xsi:schemaLocation=\"http://www.openarchives.org/OAI/2.0/oai_dc/ http://www.openarchives.org/OAI/2.0/oai_dc.xsd\">")
    current_dc.write("<title>" + new_titles[current_volume] + "</title>")
    if is_oscar:
        current_dc.write("<creator>" + oscar_project + "</creator>")
    else:    
        current_dc.write("<creator>" + project + "</creator>")
    
    if authors[current_volume] != 'null':
        current_dc.write("<creator>" + authors[current_volume] + "</creator>")
    
    contents = subjects[current_volume].split(", ")
    subject = subjects[0]
    if len(contents) > 1:
        for content in contents[1:]:
            subject += "--" + content
    
    current_dc.write("<subject>" + subject + '--' + countries[current_volume] + '--' + eras[current_volume] + "</subject>")
    
    description = abstracts[current_volume] + " " + conditions[current_volume]
    current_dc.write("<description>" + description + "</description>")
    current_dc.write("<publisher>" + publisher + "</publisher>")
    
    if countries[current_volume] == 'Cuba':
        if is_oscar:
            current_dc.write("<contributor>" + oscar_director + "</contributor>")
        else:
            current_dc.write("<contributor>" + director + "</contributor>")
            current_dc.write("<contributor>" + cuba_director + "</contributor>")
    elif countries[current_volume] == 'Brazil':
        current_dc.write("<contributor>" + director + "</contributor>")
        current_dc.write("<contributor>" + brazil_director + "</contributor>")
        
    current_dc.write("<contributor>" + metadata_creators[current_volume] + "(metadata creation)" + "</contributor>")
    current_dc.write("<source>" + sources[current_volume] + "</source>")
    current_dc.write("<identifier>" + identifiers[current_volume] + "</identifier>")
    current_dc.write("<date>" + dates[current_volume] + "</date>")    
    
    for typ in types:
        current_dc.write("<type>" + typ + "</type>")    
    
    current_dc.write("<language>" + languages[current_volume] + "</language>")
    
    current_dc.write("<coverage>" + countries[current_volume] + "</coverage>")
    current_dc.write("<coverage>" + continents[current_volume] + "--" + countries[current_volume] + "--" + states[current_volume] + "--" + cities[current_volume] + "</coverage>")
    current_dc.write("<coverage>" + coordinates[current_volume] + "</coverage>")
    current_dc.write("<coverage>" + eras[current_volume] + "</coverage>")
    current_dc.write("<rights>" + rights + "</rights>")        
    current_dc.write("</oai_dc:dc>")
    current_dc.close()
    
    print("DC completed for " + new_titles[current_volume])
    
    mods.write("<mods xmlns=\"http://www.loc.gov/mods/v3\" xmlns:mods=\"http://www.loc.gov/mods/v3\" xmlns:xsi=\"http://www.w3.org/2001/XMLSchema-instance\" xmlns:xlink=\"http://www.w3.org/1999/xlink\">")
    mods.write("<titleInfo>")
    mods.write("<title>" + new_titles[current_volume] + "</title>")
    mods.write("<subTitle/></titleInfo>")
    
    if countries[current_volume] == 'Cuba':
        if is_oscar:
            mods.write("<name type=\"corporate\"><namePart>" + oscar_project + "</namePart><role><roleTerm authority=\"marcrelator\" type=\"text\">project</roleTerm></role></name>")
            mods.write("<name type=\"personal\"><namePart>" + oscar_director[:oscar_director.find("(")] + "</namePart><role><roleTerm authority=\"marcrelator\" type=\"text\">Project Director</roleTerm></role></name>")
        else:
            mods.write("<name type=\"corporate\"><namePart>" + project + "</namePart><role><roleTerm authority=\"marcrelator\" type=\"text\">project</roleTerm></role></name>")
            mods.write("<name type=\"personal\"><namePart>" + director[:director.find("(")] + "</namePart><role><roleTerm authority=\"marcrelator\" type=\"text\">Project Director</roleTerm></role></name>")
            mods.write("<name type=\"personal\"><namePart>" + cuba_director[:cuba_director.find("(")] + "</namePart><role><roleTerm authority=\"marcrelator\" type=\"text\">Project Director</roleTerm></role></name>")
    elif countries[current_volume] == 'Brazil':
        mods.write("<name type=\"corporate\"><namePart>" + project + "</namePart><role><roleTerm authority=\"marcrelator\" type=\"text\">project</roleTerm></role></name>")
        mods.write("<name type=\"personal\"><namePart>" + director[:director.find("(")] + "</namePart><role><roleTerm authority=\"marcrelator\" type=\"text\">Project Director</roleTerm></role></name>")
        mods.write("<name type=\"personal\"><namePart>" + brazil_director[:brazil_director.find("(")] + "</namePart><role><roleTerm authority=\"marcrelator\" type=\"text\">Project Director</roleTerm></role></name>")
            
    mods.write("<name type=\"personal\"><namePart>" + metadata_creators[current_volume] + "</namePart><role><roleTerm authority=\"marcrelator\" type=\"text\">Metadata Creation</roleTerm></role></name>")
    
    mods.write("<typeOfResource>text</typeOfResource>")    
    mods.write("<genre>" + types[1] + "</genre>")
    mods.write("<tableOfContents/>")
    mods.write("<originInfo><dateIssued></dateIssued><copyrightDate/><issuance>monographic</issuance><edition/><publisher>" + publisher + "</publisher><place/></originInfo>")
    mods.write("<language><languageTerm authority=\"iso639-2b\" type=\"code\">" + languages[current_volume] + "</languageTerm></language>")    
    mods.write("<abstract>" + abstracts[current_volume] + "</abstract>")
    mods.write("<identifier type=\"essss\">" + identifiers[current_volume] + "</identifier>")
    mods.write("<physicalDescription><form>" + conditions[current_volume] + "</form><extent/></physicalDescription>")
    mods.write("<note type=\"rights statements\">" + rights + "</note>")
    mods.write("<subject><topic>" + subject + "</topic><geographic>" + sources[current_volume] + "," + cities[current_volume] + "," + states[current_volume] + "," + countries[current_volume] + "</geographic><temporal>" + dates[current_volume] + "</temporal>")
    mods.write("<hierarchicalGeographic><continent>" + continents[current_volume] + "</continent><country>" + countries[current_volume] + "</country><province>" + states[current_volume] + "</province><region/><county/><city>" + cities[current_volume] + "</city><citySection/></hierarchicalGeographic>")
    mods.write("<cartographics><coordinates>" + coordinates[current_volume] + "</coordinates></cartographics></subject>")
    mods.write("<classification/>")
    mods.write("<name/>")
    mods.write("</mods>")
    mods.close()
    
    print("MODS completed for " + new_titles[current_volume])
    
    current_volume += 1 

print("All metadata completed.")
```
