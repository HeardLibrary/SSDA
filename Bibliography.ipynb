```python
books = open("Books.tsv","r",encoding='utf-8')
bib = open("Books.txt","w",encoding='utf-8')
for line in books:
    data = line.split('\t')
    author_last = data[0]
    author_first = data[1]
    title = data[2]
    volumes = data[3]
    place_of_publication = data[4]
    press = data[5]
    publication_date = data[6]
    editor_last = data[7]
    editor_first = data[8]
    chapter_author_last = data[9]
    chapter_author_first = data[10]
    chapter_title = data[11]
    chapter_pages = data[12]
    editor_2_last = data[13]
    editor_2_first = data[14]
    author_2_last = data[15]
    author_2_first = data[16]
    editor_3_last = data[17]
    editor_3_first = data[18]
    edition = data[19]
    translator_last = data[20]
    translator_first = data[21]
    
    if (author_last != ''):
        if (volumes != ''):
            if (author_2_last != ''):
                bib.write(author_last + ', ' + author_first + ', and ' + author_2_first + ' ' + author_2_last + '. ' + title + '. ' + volumes + ' vols. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
            else:
                bib.write(author_last + ', ' + author_first + '. ' + title + '. ' + volumes + ' vols. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
        else:
            if (edition != ''):
                bib.write(author_last + ', ' + author_first + '. ' + title + '. ' + edition + ' ed. '+ place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
            else:
                if (translator_last != ''):
                    bib.write(author_last + ', ' + author_first + '. ' + title + '. Translated by ' + translator_first + ' ' + translator_last + '. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
                else:
                    if (author_2_last != ''):
                        bib.write(author_last + ', ' + author_first + ', and ' + author_2_first + ' ' + author_2_last + '. ' + title + '. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
                    else:
                        bib.write(author_last + ', ' + author_first + '. ' + title + '. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
    else:
        if (chapter_title != ''):
            if (edition != ''):
                bib.write(chapter_author_last + ', ' + chapter_author_first + '. "' + chapter_title + '." In ' + title + ', ' + edition + 'ed., edited by ' + editor_first + ' ' + editor_last + ', ' + chapter_pages + '. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
            else:
                if (editor_2_last != ''):
                    if (editor_3_last != ''):
                        bib.write(chapter_author_last + ', ' + chapter_author_first + '. "' + chapter_title + '." In ' + title + ', edited by ' + editor_first + ' ' + editor_last + ', ' + editor_2_first + ' ' + editor_2_last + ', and ' + editor_3_first + ' ' + editor_3_last + ', ' + chapter_pages + '. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
                    else:
                        bib.write(chapter_author_last + ', ' + chapter_author_first + '. "' + chapter_title + '." In ' + title + ', edited by ' + editor_first + ' ' + editor_last + ' and ' + editor_2_first + ' ' + editor_2_last + ', ' + chapter_pages + '. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
                else:
                    bib.write(chapter_author_last + ', ' + chapter_author_first + '. "' + chapter_title + '." In ' + title + ', edited by ' + editor_first + ' ' + editor_last + ', ' + chapter_pages + '. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
        else:
            if (editor_2_last != ''):
                bib.write(editor_last + ', ' + editor_first + ', and ' + editor_2_first + ' ' + editor_2_last + ', eds. ' + title + '. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
            else:
                bib.write(editor_last + ', ' + editor_first + ', ed. ' + title + '. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')

books.close()
bib.close()
```

```python
articles = open("Articles.tsv","r",encoding='utf-8')
bib = open("Articles.txt","w",encoding='utf-8')
for line in articles:
    data = line.split('\t')
    author_last = data[0]
    author_first = data[1]
    title = data[2]
    journal = data[3]
    issue = data[4]
    year = data[5]
    pages = data[6]    
    author_2_last = data[7]
    author_2_first = data[8]
    author_3_last = data[9]
    author_3_first = data[10]
    
    if (author_3_last != ''):
        bib.write(author_last + ', ' + author_first + ', ' + author_2_first + ' ' + author_2_last + ', and ' + author_3_first + ' ' + author_3_last + '. "' + title + '." ' + journal + ' ' + issue + ' (' + year + '): ' + pages + '.\n')
    elif (author_2_last != ''):
        bib.write(author_last + ', ' + author_first + ' and ' + author_2_first + ' ' + author_2_last + '. "' + title + '." ' + journal + ' ' + issue + ' (' + year + '): ' + pages + '.\n')
    else:
        bib.write(author_last + ', ' + author_first + '. "' + title + '." ' + journal + ' ' + issue + ' (' + year + '): ' + pages + '.\n')
    
articles.close()
bib.close()
```

```python
primary = open("Primary.tsv","r",encoding='utf-8')
bib = open("Primary.txt","w",encoding='utf-8')
for line in primary:
    data = line.split('\t')
    author_last = data[0]
    author_first = data[1]
    title = data[2]
    volumes = data[3]
    place_of_publication = data[4]
    press = data[5]
    publication_date = data[6]    
    editor_last = data[7]
    editor_first = data[8]
    chapter_author_last = data[9]
    chapter_author_first = data[10]
    chapter_title = data[11]
    editor_2_last = data[12]
    editor_2_first = data[13]
    
    if (author_last != ''):
        if (editor_last == ''):
            bib.write(author_last + ', ' + author_first + '. ' + title + '. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
        else:
            if (editor_2_last == ''):
                bib.write(author_last + ', ' + author_first + '. ' + title + '. Edited by ' + editor_first + ' ' + editor_last + '. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
            else:
                bib.write(author_last + ', ' + author_first + '. ' + title + '. Edited by ' + editor_first + ' ' + editor_last + ' and ' + editor_2_first + ' ' + editor_2_last + '. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
    else:
        if (editor_last == ''):
            bib.write(title + '. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
        else:
            if (volumes != ''):
                bib.write(editor_last + ', ' + editor_first + ', ed. ' + title + '. ' + volumes + ' vols. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
            else:
                if (chapter_title != ''):
                    bib.write(chapter_author_last + ', ' + chapter_author_first + '. "' + chapter_title + '." In ' + title + ', edited by ' + editor_first + ' ' + editor_last + '. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
                else:
                    if (editor_2_last == ''):
                        bib.write(editor_last + ', ' + editor_first + ', ed. ' + title + '. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
                    else:
                        bib.write(editor_last + ', ' + editor_first + ' and ' + editor_2_first + ' ' + editor_2_last + ', eds. ' + title + '. ' + place_of_publication + ': ' + press + ', ' + publication_date + '.\n')
    
primary.close()
bib.close()
```
